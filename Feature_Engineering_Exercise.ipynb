2. Transform the 'datetime' column into a datetime type and use it to create 3 new columns in the data frame containing the:
     1. Name of the Month
     2. Name of the Day of the Week
     3. Hour of the Day
        1. Make sure all 3 new columns are 'object' datatype so they can be one-hot encoded later.
        2. Drop the 'datetime' and 'season' columns. These are now redundant.

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df['Name of the Month'] = df['datetime'].dt.month_name()
df['Name of the Day of the Week'] = df['datetime'].dt.day_name()
df['Hour of the day'] = df['datetime'].dt.hour

df.head()
```

3. The temperatures in the 'temp' and 'atemp' columns are in Celsius. Use `.apply()` and a Lambda function to convert them to Fahrenheit.

```python
df['temp'] = df['temp'].apply(lambda c: 9 / 5 * c + 32)
df.head()
```

```python
df['atemp'] = df['atemp'].apply(lambda c: 9 / 5 * c + 32)
df.head()
```

Create a new column, 'temp_variance,' which shows how much warmer or colder the current temperature ('temp') is than the average temperate 
for that day of the year ('atemp'). If the current temperature is warmer than average ('atemp'), the value in 'temp_variance' should be positive.
1. Drop the 'atemp' column.

```python
avg_temp = df['atemp'].mean()
def warmer_colder(temp):
    if temp > avg_temp:    
        return 'positive'
    else:    
        return 'negative'
```

```python
df = df.assign(temp_variance=df['atemp'] * 2)
```

```python
df.head(2)
```

```python
df['temp_variance'] = df['temp_variance'].apply(warmer_colder)
df.head()
```

```python
df = df.drop(columns=[ 'atemp' ])
```
